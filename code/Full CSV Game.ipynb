```python
import pandas as pd
```

```python
sales = pd.read_csv("/Users/anish/Documents/DSP Datasets/vgsales.csv")
descriptions = pd.read_csv("/Users/anish/Documents/DSP Datasets/all_games.csv")
descriptions
```

```python
sales
```

```python
descriptions.rename(columns = {'name':'Name'}, inplace = True)
descriptions.rename(columns = {'platform':'Platform'}, inplace = True)
descriptions.rename(columns = {'release_date':'Year'}, inplace = True)
descriptions.rename(columns = {'summary':'Summary'}, inplace = True)
descriptions.head()
```

```python
platforms = descriptions['Platform'].unique()
print(platforms)
print(platforms.size)
```

```python
platforms2 = sales['Platform'].unique()
print(platforms2)
print(platforms2.size)
```

```python
descriptions['Platform'] = descriptions['Platform'].str.replace(' Nintendo 64', 'N64')

descriptions['Platform'] = descriptions['Platform'].str.replace(' PlayStation 3', 'PS3')

descriptions['Platform'] = descriptions['Platform'].str.replace(' Dreamcast', 'DC')
descriptions['Platform'] = descriptions['Platform'].str.replace(' Xbox 360', 'X360')
descriptions['Platform'] = descriptions['Platform'].str.replace(' Wii', 'Wii')

descriptions['Platform'] = descriptions['Platform'].str.replace(' Xbox One', 'XOne')
descriptions['Platform'] = descriptions['Platform'].str.replace(' PC', 'PC')
#descriptions['Platform'] = descriptions['Platform'].str.replace('Switch', '')
descriptions['Platform'] = descriptions['Platform'].str.replace(' PlayStation 2', 'PS2')

descriptions['Platform'] = descriptions['Platform'].str.replace(' PlayStation 4', 'PS4')
descriptions['Platform'] = descriptions['Platform'].str.replace(' GameCube', 'GC')
descriptions['Platform'] = descriptions['Platform'].str.replace(' Xbox', 'XB')

descriptions['Platform'] = descriptions['Platform'].str.replace(' Wii U', 'WiiU')
descriptions['Platform'] = descriptions['Platform'].str.replace(' Game Boy Advance', 'GBA')
descriptions['Platform'] = descriptions['Platform'].str.replace(' 3DS', '3DS')

#descriptions['Platform'] = descriptions['Platform'].str.replace('Xbox Series X', '')
descriptions['Platform'] = descriptions['Platform'].str.replace(' DS', 'DS')
descriptions['Platform'] = descriptions['Platform'].str.replace(' PlayStation Vita', 'PSV')

#descriptions['Platform'] = descriptions['Platform'].str.replace('PlayStation 5', '')
descriptions['Platform'] = descriptions['Platform'].str.replace(' PSP', 'PSP')
#descriptions['Platform'] = descriptions['Platform'].str.replace('Stadia', '')

descriptions['Platform'] = descriptions['Platform'].str.replace(' PlayStation', 'PS')
```

```python
platforms = descriptions['Platform'].unique()
print(platforms)
print(platforms.size)
```

```python
combine = pd.merge(sales, descriptions, on = ['Name', 'Platform'])
combine = combine[["Name","Platform","Summary", "Global_Sales"]]
combine
```

```python
count = combine['Summary'].isna().sum()
print(count)
```

```python
all_combine_dupe = combine.drop_duplicates(subset=["Name", "Platform"], keep="first", inplace=False)
all_combine_dupe
```

```python
final_data = all_combine_dupe.dropna(subset=["Summary"])
final_data
```

```python
final_data.to_csv("Games_Dataset_Model.csv")
```
